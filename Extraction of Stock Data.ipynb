```python
# Using yfinance to Extract Tesla Stock Data with TSLA as ticker symbol

tsla = yf.Ticker('TSLA')
```

Using the ticker object and the function `history` to extract stock information and saved in a dataframe named `tesla_data`. The `period` parameter is set to ` "max" ` to get information for the maximum amount of time.


```python
tesla_data = tsla.history(period="max")
```

```python
# index reset

tesla_data = tsla.history(period='max')
tesla_data.head()
```

```python
tesla_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm'
```

By using Webscraping to Extract Tesla Revenue Data, the `request` library is used to download the `tesla_url` webpage and save the text of the response as a variable named `tesla_data`

```python
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

The read_html function is used to extract the table

```python
# Using read_html function extract the table with Tesla Revenue
read_html_pandas_data = pd.read_html(tesla_url)
tesla_revenue = read_html_pandas_data[0]

tesla_revenue.head()
```

```python
# The columns are renamed to date and revenue

tesla_revenue = tesla_revenue.rename(columns={'Tesla Annual Revenue (Millions of US $)': 'Date', 'Tesla Annual Revenue (Millions of US $).1': 'Revenue'})
tesla_revenue.head()
```

```python
# The comma and dollar sign are removed from the Revenue column

tesla_revenue["Revenue"] = tesla_revenue['Revenue'].str.replace('$',"")
tesla_revenue.head()
```
